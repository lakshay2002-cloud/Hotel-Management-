```python
import numpy as np 
class HotelBilling:

    def __init__(self):
        print("\n\n*****WELCOME TO HEWING HOTEL*****\n")
        self.room_rent = 0
        self.food_bill = 0
        self.laundry_bill = 0
        self.game_bill = 0
        self.service_charge = 1800
        self.customer_name = ''
        self.customer_address = ''
        self.check_in_date = ''
        self.check_out_date = ''
        self.room_no = np.random.randint(100, 1000)

    def input_customer_data(self):
        self.customer_name = input("Enter your name: ")
        self.customer_address = input("Enter your address: ")
        self.check_in_date = input("Enter your check-in date: ")
        self.check_out_date = input("Enter your check-out date: ")
        print("Your room no.:", self.room_no, "\n")

    def calculate_room_rent(self):
        print("We have the following rooms for you:")
        print("1. Honeymoon Suite   --> Rs 10000 per night")
        print("2. Presidential Suite  --> Rs 15000 per night")
        print("3. Suite --> Rs 6000 per night")
        print("4. Junior Suit --> Rs 5000 per night")
        print("5. Studio Room   --> Rs 5500 per night")
        print("6. Penthouse   --> Rs 9000 per night")
        print("7. Double Bed --> Rs 4000 per night")
        print("8. Single Bed  --> Rs 3000 per night")
        
        room_type = int(input("Enter your choice: "))
        nights = int(input("How many nights did you stay? "))

        rates = {1: 10000, 2: 15000, 3: 6000, 4: 5000, 5: 5500, 6: 9000, 7: 4000, 8: 3000}
        self.room_rent = rates.get(room_type, 0) * nights

        if self.room_rent == 0:
            print("Invalid room type selected!")
        else:
            print("Your room rent is: Rs","",self.room_rent, "\n")

    def calculate_restaurant_bill(self):
        print("*****RESTAURANT MENU*****")
        print("1. Water --> Rs 20")
        print("2. Tea --> Rs 10")
        print("3. Breakfast combo --> Rs 90")
        print("4. Lunch --> Rs 110")
        print("5. Dinner --> Rs 150")
        print("6. Coffee --> Rs 40")
        print("7. Sandwich --> Rs 50")
        print("8. Dessert --> Rs 80")
        print("9. Exit")

        prices = {1: 20, 2: 10, 3: 90, 4: 110, 5: 150, 6: 40, 7: 50, 8: 80}

        while True:
            choice = int(input("Enter your choice: "))
            if choice == 9:
                break
            elif choice in prices:
                quantity = int(input("Enter the quantity: "))
                self.food_bill += prices[choice] * quantity
            else:
                print("Invalid option")

        print("Total food cost: Rs", self.food_bill, "\n")

    def calculate_laundry_bill(self):
        print("*****LAUNDRY MENU*****")
        print("1. Shorts --> Rs 3")
        print("2. Trousers --> Rs 4")
        print("3. Shirt --> Rs 5")
        print("4. Jeans --> Rs 6")
        print("5. Girlsuit --> Rs 8")
        print("6. Skirt --> Rs 7")
        print("7. Jacket --> Rs 10")
        print("8. Blazer --> Rs 12")
        print("9. Exit")

        prices = {1: 3, 2: 4, 3: 5, 4: 6, 5: 8, 6: 7, 7: 10, 8: 12}

        while True:
            choice = int(input("Enter your choice: "))
            if choice == 9:
                break
            elif choice in prices:
                quantity = int(input("Enter the quantity: "))
                self.laundry_bill += prices[choice] * quantity
            else:
                print("Invalid option")

        print("Total laundry cost: Rs", self.laundry_bill, "\n")

    def calculate_game_bill(self):
        print("*****GAME MENU*****")
        print("1. Table tennis --> Rs 60 per hour")
        print("2. Bowling --> Rs 80 per hour")
        print("3. Snooker --> Rs 70 per hour")
        print("4. Video games --> Rs 90 per hour")
        print("5. Pool --> Rs 50 per hour")
        print("6. Chess --> Rs 30 per hour")
        print("7. Carrom --> Rs 40 per hour")
        print("8. Exit")

        prices = {1: 60, 2: 80, 3: 70, 4: 90, 5: 50, 6: 30, 7: 40}

        while True:
            choice = int(input("Enter your choice: "))
            if choice == 8:
                break
            elif choice in prices:
                hours = int(input("Enter the number of hours: "))
                self.game_bill += prices[choice] * hours
            else:
                print("Invalid option")

        print("Total game cost: Rs", self.game_bill, "\n")

    def display_bill(self):
        print("\n******HOTEL BILL******")
        print("Customer details:")
        print("Name:", self.customer_name)
        print("Address:", self.customer_address)
        print("Check-in date:", self.check_in_date)
        print("Check-out date:", self.check_out_date)
        print("Room no.:", self.room_no)
        print("Room rent: Rs", self.room_rent)
        print("Food bill: Rs", self.food_bill)
        print("Laundry bill: Rs", self.laundry_bill)
        print("Game bill: Rs", self.game_bill)

        total = self.room_rent + self.food_bill + self.laundry_bill + self.game_bill
        print("Subtotal: Rs", total)
        print("Additional service charges: Rs", self.service_charge)
        print("Grand total: Rs", total + self.service_charge, "\n")
        self.room_no += 1

def main():
    hotel = HotelBilling()

    while True:
        print("1. Enter customer data")
        print("2. Calculate room rent")
        print("3. Calculate restaurant bill")
        print("4. Calculate laundry bill")
        print("5. Calculate game bill")
        print("6. Show total cost")
        print("7. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            hotel.input_customer_data()
        elif choice == 2:
            hotel.calculate_room_rent()
        elif choice == 3:
            hotel.calculate_restaurant_bill()
        elif choice == 4:
            hotel.calculate_laundry_bill()
        elif choice == 5:
            hotel.calculate_game_bill()
        elif choice == 6:
            hotel.display_bill()
        elif choice == 7:
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
